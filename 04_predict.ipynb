```python
import tensorflow as tf
from tensorflow.keras.models import load_model
import pickle
import pandas as pd
import numpy as np
```

```python
# Load the trained model and preprocessing pkl files
model=load_model('ann_model.h5')

## load the encoder and scaler
with open('oh_encoder_geo.pkl','rb') as file:
    oh_encoder_geo=pickle.load(file)

with open('label_encoder_gender.pkl', 'rb') as file:
    label_encoder_gender = pickle.load(file)

with open('scaler.pkl', 'rb') as file:
    scaler = pickle.load(file)
```

```python
# Example input data
input_data = {
    'CreditScore': 600,
    'Geography': 'France',
    'Gender': 'Male',
    'Age': 40,
    'Tenure': 3,
    'Balance': 60000,
    'NumOfProducts': 2,
    'HasCrCard': 1,
    'IsActiveMember': 1,
    'EstimatedSalary': 50000
}
```

```python
input_df = pd.DataFrame([input_data])
```

```python
input_df = label_encoder_gender.transform(input_df)
```

```python
input_df
```

```python
input_df = oh_encoder_geo.transform(input_df)
```

```python
input_df
```

```python
input_df = pd.DataFrame(
    scaler.transform(input_df),
    columns=input_df.columns
)
```

```python
input_df
```

```python
## Predict churn
prediction=model.predict(input_df)
prediction
```

```python
prediction_proba = prediction[0][0]
prediction_proba
```

```python
if prediction_proba > 0.5:
    print('The customer is likely to churn.')
else:
    print('The customer is not likely to churn.')
```
